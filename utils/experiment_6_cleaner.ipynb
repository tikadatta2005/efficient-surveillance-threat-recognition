### Cleaner Code.

This code cleans and merge the file extracted in traing for experiment_6

```python
import pandas as pd
```

```python
train_metrics = [
    {"path":"../documentations/experiments/experiment6/tables/training_metrics.csv", "start":1, "end":100},
    {"path":"../documentations/experiments/experiment6/tables/training_metrics_200.csv", "start":101, "end":200},
    {"path":"../documentations/experiments/experiment6/tables/training_metrics_400.csv", "start":201, "end":400}
]

dfs = []

for obj in train_metrics:
    df = pd.read_csv(obj["path"])
    df["epoch"] = range(obj["start"], obj["end"] + 1)
    dfs.append(df)

training_df = pd.concat(dfs, ignore_index=True)
len(training_df)
```

```python
training_df.to_csv("../documentations/experiments/experiment6/tables/training_metrics.csv", index=False, mode="w")
```

```python
test_metrics = [
    {"path":"../documentations/experiments/experiment6/tables/test_metrics.csv", "sum":0},
    {"path":"../documentations/experiments/experiment6/tables/test_metrics_200.csv", "sum":100},
    {"path":"../documentations/experiments/experiment6/tables/test_metrics_400.csv", "sum":200}
]

dfs = []

for obj in test_metrics:
    df = pd.read_csv(obj["path"])
    df["epoch"] = df["epoch"] + obj["sum"]
    dfs.append(df)
    
test_df = pd.concat(dfs, ignore_index=True)
len(test_df)
```

```python
test_df.to_csv("../documentations/experiments/experiment6/tables/test_metrics.csv", index=False, mode="w")
```
